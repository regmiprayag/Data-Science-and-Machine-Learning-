```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("fraud_0.1origbase.csv")
```

```python
data.head()
```

```python
data.isFraud.value_counts()
```

```python
data.columns
```

```python
data.drop(['step','nameOrig','nameDest'], inplace = True, axis =1)
```

```python
data
```

```python
data_non_fraud = data[data['isFraud']==0][:1000]
```

```python
data_fraud = data[data['isFraud']==1]
```

```python
data_non_fraud
```

```python
data_fraud
```

```python
data = pd.concat([data_non_fraud, data_fraud])
```

```python
data
```

```python
label_enc = LabelEncoder()
```

```python
data['type'] = label_enc.fit_transform(data['type'])
```

```python
data
```

```python
data.dtypes
```

```python
X = data.drop('isFraud', axis=1)
y = data['isFraud']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, stratify=y, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_class = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_class.fit(X_train, y_train)
```

```python
y_pred = knn_class.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy = accuracy_score(y_pred, y_test.values)
```

```python
accuracy
```

```python
class_report = classification_report(y_pred, y_test.values)
```

```python
print(class_report)
```

```python
conf_mat = confusion_matrix(y_pred, y_test.values)
```
